```python
# 미션 진행을 위해 필요한 라이브러리를 로드합니다.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://bit.ly/telco-csv", index_col="customerID")
df.shape
```

```python
#Return Series with number of distinct elements. Can ignore NaN values.
df['tenure']
```

```python
import matplotlib.pyplot as plt

plt.hist(df['tenure'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of tenure')
plt.xlabel('tenure')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=df['tenure'])
plt.title('Box plot of tenure')
plt.show()
```

```python
unique_tenure_values = df['tenure'].unique()
print("Unique values in tenure:", unique_tenure_values)
```

```python
print(df['tenure'].describe())
```

```python
sns.kdeplot(df['tenure'], fill=True, color='skyblue')
plt.title('Kernel Density Estimation of tenure')
plt.xlabel('tenure')
plt.ylabel('Density')
plt.show()
```

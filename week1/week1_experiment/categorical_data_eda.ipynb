# 상관관계 분석하기

```python
df["TotalCharges"] = df["TotalCharges"].str.strip().replace("", np.nan).astype(float)
```

```python
df = df.dropna()
```

```python
df["gender_encode"] = (df["gender"] == "Male").astype(int) #male -> 1 female -> 0
df["Partner_encode"] = (df["Partner"] == "Yes").astype(int)
df["Dependents_encode"] = (df["Dependents"] == "Yes").astype(int)
df["PhoneService_encode"] = (df["PhoneService"] == "Yes").astype(int)
df["PaperlessBilling_encode"] = (df["PaperlessBilling"] == "Yes").astype(int)
```

```python
df["gender_encode"]
```

```python
feature_names = df.select_dtypes(include='number').columns
feature_names
```

```python
df["Churn_encode"] = (df["Churn"] == "Yes").astype(int)
```

```python
# df는 데이터프레임, 'churn'은 분류할 값
correlation_matrix = df[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges',
       'gender_encode', 'Partner_encode', 'Dependents_encode',
       'PhoneService_encode', 'PaperlessBilling_encode','Churn_encode']].corr()

correlation_matrix
churn_correlation = correlation_matrix['Churn_encode'].sort_values(ascending=False)
churn_correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# heatmap을 이용한 상관관계 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 상관관계의 절댓값이 0.2보다 큰 특성들을 선택
significant_correlations = churn_correlation[churn_correlation > 0.1].index

# 유의미한 상관관계를 가지는 요소들로 새로운 상관계수 행렬을 생성
significant_correlation_matrix = df[significant_correlations].corr()

# heatmap을 이용한 상관관계 시각화
plt.figure(figsize=(6, 4))
sns.heatmap(significant_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

# 범주형 값들을 적절히 변환하여 상관관계 분석하기

```python
# df는 데이터프레임
column_data_types = df.dtypes

# 각 칼럼의 데이터 타입 출력
print(column_data_types)
```

```python
selected_columns = df.columns[df.apply(lambda x: x.nunique()) == 3]

# 선택된 열 출력
print(selected_columns)
```

```python
# 선택된 열들을 이진(dummy) 변수로 변환
df = pd.get_dummies(df, columns=selected_columns)
```

```python
df.head(1)
```

```python
df.replace({'Yes': 1, 'No': 0}, inplace=True)
df.replace({False: 0, True: 1}, inplace=True)
df.head(1)
```

```python
numeric_columns = df.select_dtypes(include=['int', 'float']).columns
numeric_columns
```

```python
# df는 데이터프레임, 'churn'은 분류할 값
correlation_matrix = df[numeric_columns].corr()

correlation_matrix
churn_correlation = correlation_matrix['Churn_encode'].sort_values(ascending=False)
churn_correlation
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 상관관계의 절댓값이 0.2보다 큰 특성들을 선택
significant_correlations = churn_correlation[churn_correlation > 0.3].index

# 유의미한 상관관계를 가지는 요소들로 새로운 상관계수 행렬을 생성
significant_correlation_matrix = df[significant_correlations].corr()

# heatmap을 이용한 상관관계 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(significant_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```
